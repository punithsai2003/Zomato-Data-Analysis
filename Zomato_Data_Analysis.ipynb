```python
# Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create Data Frame
Dataframe = pd.read_csv('/content/Zomato data .csv')
Dataframe
```

**Data Cleaning**


```python
# convert the rate column

def convert_rate(value):
  value = str(value).split('/')
  value = value[0];
  return float(value)

Dataframe['rate'] = Dataframe['rate'].apply(convert_rate)
Dataframe
```

```python
Dataframe.info()
```

```python
Dataframe.head()
```

**1) What Type of Restaurants do Majority of customers Order From ?**

```python
#Type of Restaurant
#plt.figure(figsize=(10,6))
sns.countplot(x=Dataframe['listed_in(type)'])
plt.xlabel("Type of Restaurant")
```

**Conclusion - Majority of restaurants falls in dining category**

**2) How many Votes has each type of Restaurant recieved from Customers?**


```python
Dataframe.head()
```

```python
Grouped_data = Dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': Grouped_data})
plt.plot(result, c="green",marker="o")
plt.xlabel("Type of Restaurant", c="red", size=15)
plt.ylabel("Votes", c="red", size=15)
```

**Conclusion - Dining Restaurants Recieved maximum number of Votes**

**3) What are Ratings that the majority of Restaurants have Recieved**

```python
Dataframe.head()
```

```python
plt.hist(Dataframe['rate'],bins=5)
plt.title ("Rating Distribution")
plt.xlabel("Rating")
plt.show()
```

**Conclusion -  Majority of Restaurants have recieved ratings from 3.5 to 4**

**4) Zomato has observed that most couples order most of thier food online. What is thier average Spending on each order ?**

```python
Dataframe.head()
```

```python
couple_data = Dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
plt.show()
```

**Conclusion - The couples spend average of 300 rupees per order**

**5) Which mode has recieved maximum Rating ?**

```python
Dataframe.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x="online_order", y='rate', data = Dataframe)
```

**Conclusion -  Online Orders has recieved Maximum ratings in comparison to Offline orders**

***6)Which type of restaurant recieved more offline orders, so that Zomato can provide Customers with good offers? ***

```python
Dataframe.head()
```

```python
pivot_table = Dataframe.pivot_table(index='listed_in(type)', columns='online_order',aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("listed_in(type)")
plt.show()
```

**Conclusion - Dining type of restuarants has recieved more offline orders**
